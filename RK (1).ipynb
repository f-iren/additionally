# Реализация метода Рунге-Кутта

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams['font.size'] = '16'
```

```python
N=20
al=1/2
a=0
b=1
l=b-a
h=l/N
y0=0
```

```python
def f(t):
    return sin(t*t+t)
```

```python
def integ(x, f):
    N = 1
    h = x
    cur = 0.0
    next = 0.0
    eps=0.00000001
    next = (f(0)+f(x))/2
    while(1):
        cur = next
        h = h / 2;
        for i in range(N):
            next += f((i*2+1)*h)
        N = N * 2
        if(fabs(h*2*cur-h*next)<eps):
           return(h*next)
```

```python
#Функция в правой части дифференциального уравнения
def p(x):
    return integ(x,f)
```

```python
#Функция для нахождения следующего значения
def next_y(y,i,al,X):
    REZ=y+((1-al)*p(X[i-1])+al*p((X[i-1])+h/(2*al)))*h
    #print("rez [",i,"]=",REZ)
    return REZ
```

```python
#Реализация метода Рунге-Кутта
def RK(al,N,X,Y):
    h=l/N
    #Y=np.array([next_y(,i,al) for i in range(N+1)])
    Y[0]=y0
    cur_y=Y[0]
    for i in range(1,N+1):
        Y[i]=next_y(cur_y,i,al,X) 
        #print("Y[",i,"]=",Y[i],"\n")
        cur_y=Y[i]
    #print(Y)
```

```python
def g(x):
    return integ(x,p)
```
